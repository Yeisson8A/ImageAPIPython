```python
import requests
from PIL import Image
import urllib.request
import matplotlib.pyplot as plt
```

```python
url = "https://api.monsterapi.ai/v1/generate/txt2img"
api_key = "XXXXXXXXXXXXXXXXXXXXXXX"
```

```python
payload = {
    "safe_filter": True,
    "aspect_ratio": "square",
    "prompt": "a cat in an surf table"
}
headers = {
    "accept": "application/json",
    "content-type": "application/json",
    "authorization": "Bearer " + api_key
}
```

```python
response = requests.post(url, json=payload, headers=headers)
data = response.json()
```

```python
url_response = data['status_url']
print(url_response)
```

```python
headers = {
    "accept": "application/json",
    "authorization": "Bearer " + api_key
}
```

```python
response = requests.get(url_response, headers=headers)
data = response.json()
```

```python
result = data['result']

for image in result['output']:
    with urllib.request.urlopen(image) as url:
      img = Image.open(url)
      plt.imshow(img)
      plt.axis('off')
      plt.show()
```
